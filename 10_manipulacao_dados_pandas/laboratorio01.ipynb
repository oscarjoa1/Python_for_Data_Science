> 2. Determine a taxa de ocorrência de tuberculose para cada 10.000 pessoas. Armazene o resultado em um objeto chamado taxas .

```python
# usando table1
table1['taxas'] = table1['cases'] / table1['population'] * 10000
table1
```

```python
# usando table2
table2 = table2.pivot(index = ['country', 'year'], columns = 'type', values = 'count').reset_index().rename_axis('', axis = 1)
table2['taxas'] = table2['cases'] / table2['population'] * 10000
table2


```

```python
# usando table3
table3[ ['cases', 'population'] ] = table3['rate'].str.split(pat = "/", expand = True)
table3 = table3.drop('rate', axis = 1)
table3['taxas'] = table3['cases'].astype(float) / table3['population'].astype(float) * 10000
table3
```

```python
# usando table4a
table4a = table4a.melt(id_vars = ['country'], value_vars = ['1999', '2000'], var_name = 'year', value_name = 'cases')
table4a
```

```python
# usando table4b
table4b = table4b.melt(id_vars = ['country'], value_vars = ['1999', '2000'], var_name = 'year', value_name = 'population')
table4b
```
